```python
import pandas as pd
df=pd.read_csv("/content/archive (10).zip")
df
df.isnull().sum()

#fill with mode
df['P'].fillna(df['P'].median(),inplace=True)
df['Air temperature (C)'].fillna(df['Air temperature (C)'].median(),inplace=True)
df['Wind speed (Km/h)'].fillna(df['Wind speed (Km/h)'].median(),inplace=True)
df['Air humidity (%)'].fillna(df['Air humidity (%)'].median(),inplace=True)
df['Wind gust (Km/h)'].fillna(df['Wind gust (Km/h)'].median(),inplace=True)
df['Pressure (KPa)'].fillna(df['Pressure (KPa)'].median(),inplace=True)
df['ph'].fillna(df['ph'].median(),inplace=True)
df['rainfall'].fillna(df['rainfall'].median(),inplace=True)
df['N'].fillna(df['N'].median(),inplace=True)
df['K'].fillna(df['K'].median(),inplace=True)

# find unique values

for col in df.columns:
  print(f"Unique values for column '{col}': {df[col].unique()}")

#remove outliers

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
    return data



# Apply the function to each column you want to process
for col in ['N', 'P', 'K', 'Air temperature (C)', 'Air humidity (%)',
            'Wind speed (Km/h)', 'Wind gust (Km/h)', 'Pressure (KPa)', 'ph', 'rainfall']:
    df = remove_outliers_iqr(df, col)

df


#visualize in box plot

import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'df' is your DataFrame and you want box plots for numerical columns
numerical_cols = ['P', 'Air temperature (C)', 'Wind speed (Km/h)', 'Air humidity (%)',
                  'Wind gust (Km/h)', 'Pressure (KPa)', 'ph', 'rainfall', 'N', 'K']

for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=df[col])
    plt.title(f'Box Plot of {col}')
    plt.show()

# visualize in box plot
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'df' is your DataFrame and you want box plots for numerical columns
numerical_cols = ['P', 'Air temperature (C)', 'Wind speed (Km/h)', 'Air humidity (%)',
                  'Wind gust (Km/h)', 'Pressure (KPa)', 'ph', 'rainfall', 'N', 'K']

for col in numerical_cols:
    plt.figure(figsize=(10, 10))
    sns.boxplot(x=df[col])
    plt.title(f'Box Plot of {col}')
    plt.show()

# convert to numerical value
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
df['Status'] = labelencoder.fit_transform(df['Status'])


from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
# Assuming 'df' is your DataFrame and the features are already selected
features_to_scale = [
'Soil Moisture',	'Temperature',	'Time'	,'Air temperature (C)',	'Wind speed (Km/h)',	'Air humidity (%)'	,'Wind gust (Km/h)',	'Pressure (KPa)',	'ph'	,'rainfall'	,'N',	'P'	,'K'	,'Status'
]
# Standardization (using StandardScaler)
scaler = StandardScaler()
df[[
  'Soil Moisture',	'Temperature'	,	'Time'	,'Air temperature (C)',	'Wind speed (Km/h)',	'Air humidity (%)'	,'Wind gust (Km/h)',	'Pressure (KPa)',	'ph'	,'rainfall'	,'N',	'P'	,'K'	,'Status'
]] = scaler.fit_transform(df[features_to_scale])

# Min-Max scaling (using MinMaxScaler)
minmax_scaler = MinMaxScaler()
df[['Soil Moisture',	'Temperature'	,	'Time'	,'Air temperature (C)',	'Wind speed (Km/h)',	'Air humidity (%)'	,'Wind gust (Km/h)',	'Pressure (KPa)',	'ph'	,'rainfall'	,'N',	'P'	,'K'	,'Status'
]] = minmax_scaler.fit_transform(df[features_to_scale])

# Print the scaled values (optional)
print(df[[
'Soil Moisture',	'Temperature',	'Time'	,'Air temperature (C)',	'Wind speed (Km/h)',	'Air humidity (%)'	,'Wind gust (Km/h)',	'Pressure (KPa)',	'ph'	,'rainfall'	,'N',	'P'	,'K'	,'Status'
]])


import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error

# Assuming 'Soil Moisture' is the target variable now
# since 'Soil Humidity' has been dropped
X = df[['Temperature',	'Time'	,'Air temperature (C)',	'Wind speed (Km/h)',	'Air humidity (%)'	,'Wind gust (Km/h)',	'Pressure (KPa)',	'ph'	,'rainfall'	,'N',	'P'	,'K'	,'Status']]
y = df["Soil Moisture"]  # Changed target variable to 'Soil Moisture'

# One-hot encode categorical variables (if any remain in X)
X_encoded = pd.get_dummies(X, drop_first=True)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test data
y_pred = model.predict(X_test)

# Evaluate the model
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f'Mean Absolute Error: {mae:.2f}')
print(f'R-squared: {r2:.2f}')
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse:.2f}")


# Summarize key findings
import pandas as pd
summary = {'Dataset Shape': df.shape,
    'Missing Values': df.isnull().sum().sum(),
    'Correlation with soil measure': df.corr()['Soil Moisture'].sort_values(ascending=False)
}
print("EDA Summary:")
for key, value in summary.items():
    print(f"{key}: {value}")
# Prepare final dataset for modeling

final_df = df[['Temperature',	'Time'	,'Air temperature (C)',	'Wind speed (Km/h)',	'Air humidity (%)'	,'Wind gust (Km/h)',	'Pressure (KPa)',	'ph'	,'rainfall'	,'N',	'P'	,'K'	,'Status']]
```
